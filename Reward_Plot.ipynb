```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

# df = pd.read_csv("1_PPO_MeanAndStdReward.csv")
# my_list = df["Mean Rewards"].values
# my_list
```

```python
def instant_switch(y_ppo, y_ppo_error, y_dqn, y_dqn_error, y_greedy, y_greedy_error):

    x = [100,200,300,400,500,600,700,800,900,1000]

    y_max = [2,2,2,2,2,2,2,2,2,2]

    plt.figure(figsize=(15,10))
    y_ppo_neg_error = [y_ppo - y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
    y_ppo_pos_error = [y_ppo + y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
    y_dqn_neg_error = [y_dqn - y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
    y_dqn_pos_error = [y_dqn + y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
    y_greedy_neg_error = [y_greedy - y_greedy_error for y_greedy, y_greedy_error in zip(y_greedy, y_greedy_error)]
    y_greedy_pos_error = [y_greedy + y_greedy_error for y_greedy, y_greedy_error in zip(y_greedy, y_greedy_error)]
    
    plt.plot(x, y_ppo, 'b-', label='PPO')
    plt.plot(x, y_dqn, 'c-', label = 'DQN')
    plt.plot(x, y_greedy, 'm-', label = 'Greedy')
    plt.plot(x, y_max, 'y-', label = 'Maximum Reward')
    plt.xlabel("Episodes", fontsize =12)
    plt.ylabel("Cumulative Reward", fontsize =12)
    plt.title("Agents switching their target instantly", fontsize = 12)
    plt.legend(loc="upper left")
    plt.xticks(np.arange(1100,step=100))
    plt.ylim(bottom=0, top=2.3)
    plt.fill_between(x, y_ppo_neg_error, y_ppo_pos_error , alpha=0.4)
    plt.fill_between(x, y_greedy_neg_error, y_greedy_pos_error, alpha = 0.4)
    plt.fill_between(x, y_dqn_neg_error, y_dqn_pos_error, alpha=0.4)
    plt.show()
```

```python
def no_queuing(y_ppo, y_ppo_error, y_dqn, y_dqn_error, y_greedy, y_greedy_error):

    x = [100,200,300,400,500,600,700,800,900,1000]

    y_max = [2,2,2,2,2,2,2,2,2,2]

    plt.figure(figsize=(15,10))
    y_ppo_neg_error = [y_ppo - y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
    y_ppo_pos_error = [y_ppo + y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
    y_dqn_neg_error = [y_dqn - y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
    y_dqn_pos_error = [y_dqn + y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
    y_greedy_neg_error = [y_greedy - y_greedy_error for y_greedy, y_greedy_error in zip(y_greedy, y_greedy_error)]
    y_greedy_pos_error = [y_greedy + y_greedy_error for y_greedy, y_greedy_error in zip(y_greedy, y_greedy_error)]

    plt.plot(x, y_ppo, 'b-', label='PPO')
    plt.plot(x, y_dqn, 'c-', label = 'DQN')
    plt.plot(x, y_greedy, 'm-', label = 'Greedy')
    plt.plot(x, y_max, 'y-', label = 'Maximum Reward')
    plt.xlabel("Episodes", fontsize =12)
    plt.ylabel("Cumulative Reward", fontsize=12)
    plt.title("Agents ignoring commands until they reach their destination", fontsize=12)
    plt.legend(loc="upper left")
    plt.xticks(np.arange(1100,step=100))
    plt.ylim(bottom=0, top=2.3)
    plt.fill_between(x, y_ppo_neg_error, y_ppo_pos_error , alpha=0.4)
    plt.fill_between(x, y_greedy_neg_error, y_greedy_pos_error, alpha = 0.4)
    plt.fill_between(x, y_dqn_neg_error, y_dqn_pos_error, alpha=0.4)
    plt.show()
```
